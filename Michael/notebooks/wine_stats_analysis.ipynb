```python
# import dependancies
import os
import pandas as pd
import hvplot.pandas
import matplotlib.pyplot as plt
```

```python
# Define the paths for all source files
cwd = os.getcwd()
orig_test_data_path = os.path.join(cwd,'wine_data','test.csv')
orig_train_data_path = os.path.join(cwd,'wine_data','train.csv')
```

```python
# Open all source files as a pd dataframe
orig_test = pd.read_csv(orig_test_data_path)
orig_train = pd.read_csv(orig_train_data_path)
red_data = pd.read_csv('red_data_noID.csv')
white_data = pd.read_csv('white_data_noID.csv')
red_white_data = pd.read_csv('red_white_data_noID.csv')
trainable_data = pd.read_csv('trainable_data_noID.csv')

# Drop 'ID' columns
orig_test = orig_test.drop(columns='Id')
orig_train = orig_train.drop(columns='Id')
```

```python
acid_metrics = ['fixed acidity', 'volatile acidity', 'citric acid','pH','chlorides']
sulfur_metrics = ['free sulfur dioxide', 'total sulfur dioxide','sulphates']
texture_metrics = ['residual sugar','density','alcohol', 'quality']
```

```python
box_orig_train = orig_train.hvplot.box(invert=True,by='Category', title="Original Train Data")
box_red_data = red_data.hvplot.box(invert=True,title="Red Wine")
box_white_data = white_data.hvplot.box(invert=True,title="White Wine")
box_red_white_data = red_white_data.hvplot.box(invert=True,title="Red & White Data")
box_trainable_data = trainable_data.hvplot.box(invert=True,title="All data combined")
(box_orig_train + box_red_data + box_white_data + box_red_white_data + box_trainable_data).cols(2)
```
